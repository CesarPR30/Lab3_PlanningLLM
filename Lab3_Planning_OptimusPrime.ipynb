```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, TextStreamer
import torch
```

```python
MODEL_ID = "Qwen/Qwen3-8B" # "Qwen/Qwen3-4B"
```

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
)

tokenizer = AutoTokenizer.from_pretrained(
    MODEL_ID,
    use_fast=True,
    trust_remote_code=True
)

model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.float16,
    trust_remote_code=True
)
model.eval()

#                                  -------------------------------
#           ---------             | --------                  --- |            -----------
# texto -> |Tokenizer| - > IDs -> ||Codebook| -> Vectores -> |LLM|| -> IDs -> |unTokenizer| -> Texto (output)
#           ---------             | --------                  --- |            -----------
#                                  -------------------------------
```

```python
def qwen( prompt:str,
          system:str|None=None,
          max_new_tokens:int=512,
          temperature:float=0.8,
          top_p:float=0.9,
          enable_thinking:bool=False,
          do_sample=True,
          stream:bool=False) -> str:
    # Alta temperatura: Creativo (alucinar)
    #                   Las probabilidades eran homogeneas
    # Menor temperatura: M√°s determinista
    #                    Se aproxima a un one-hot

    # top-p: Los token m√°s probables hasta tener p de probabilidad
    # top-k: Seleccionamos los k tokens mas probables

    messages = []
    if system:
        messages.append({"role": "system", "content": system})
    messages.append({"role": "user", "content": prompt})

    # ID de tokens (respuesta)
    text = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        tokenize=False,
        enable_thinking=enable_thinking,
    )
    # Traducimos los Ids como texto
    inputs = tokenizer([text], return_tensors="pt").to(model.device)

    gen_kwargs = dict(
        max_new_tokens=max_new_tokens,
        do_sample=do_sample,
        temperature=temperature,
        top_p=top_p,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.eos_token_id,
    )

    if stream:
        streamer = TextStreamer(tokenizer, skip_prompt=True, skip_special_tokens=True)
        _ = model.generate(**inputs, streamer=streamer, **gen_kwargs)
        return ""


    with torch.no_grad():
        # resp = [input, output]
        out = model.generate(**inputs, **gen_kwargs)
    # S√≥lo la parte nueva:
    gen_ids = out[0, inputs.input_ids.shape[1]:]
    return tokenizer.decode(gen_ids, skip_special_tokens=True)
```

```python
print(qwen(
    system="Eres un docente y respondes con precisi√≥n y concisi√≥n.",
    prompt="Cuentame que paso en el Jurasico.",
    max_new_tokens=500,
    temperature=0.01,
    top_p=0.9,
    enable_thinking=False,
    stream=True
))
```

# Chatbot

```python
class Chatbot:
    def __init__(self,
                 system="Eres un asistente √∫til y preciso.",
                 temperature=0.8,
                 top_p=0.9,
                 max_new_tokens=512,
                 enable_thinking=False,
                 stream=False):
        self.system = system
        self.temperature = temperature
        self.top_p = top_p
        self.max_new_tokens = max_new_tokens
        self.enable_thinking = enable_thinking
        self.stream = stream
        self.messages = []
        if self.system:
            self.messages.append({"role": "system", "content": self.system})

    def reset(self, new_system=None):
        self.messages = []
        if new_system is not None:
            self.system = new_system
        if self.system:
            self.messages.append({"role": "system", "content": self.system})

    @torch.inference_mode()
    def ask(self, user_text):
        self.messages.append({"role": "user", "content": user_text})

        chat_text = tokenizer.apply_chat_template(
            self.messages,
            add_generation_prompt=True,
            tokenize=False,
            enable_thinking=self.enable_thinking,
        )
        inputs = tokenizer([chat_text], return_tensors="pt").to(model.device)

        gen_kwargs = dict(
            max_new_tokens=self.max_new_tokens,
            do_sample=True,
            temperature=self.temperature,
            top_p=self.top_p,
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.eos_token_id,
        )

        if self.stream:
            streamer = TextStreamer(tokenizer, skip_prompt=True, skip_special_tokens=True)
            _ = model.generate(**inputs, streamer=streamer, **gen_kwargs)
            out_ids = model.generate(**inputs, **gen_kwargs)
        else:
            out_ids = model.generate(**inputs, **gen_kwargs)

        gen_ids = out_ids[0, inputs.input_ids.shape[1]:]
        reply = tokenizer.decode(gen_ids, skip_special_tokens=True).strip()

        self.messages.append({"role": "assistant", "content": reply})
        return reply
```

```python
bot = Chatbot(
    system="Eres un ni√±o de 3 a√±os.",
    temperature=0.7,
    top_p=0.1,
    max_new_tokens=1024,
    enable_thinking=False,
    stream=True
)
```

```python
import json
from pathlib import Path

EXAMPLES_PATH = Path("/content/Examples.json")
TASK_PATH = Path("/content/Task.json")
OUT_PATH = Path("/content/submission.json")

def load_json(path):
    return json.loads(path.read_text(encoding="utf-8"))

def split_examples(examples):
    blocks, objects = [], []
    for ex in examples:
        sc = ex["scenario_context"]
        if "set of blocks" in sc:
            blocks.append(ex)
        else:
            objects.append(ex)
    return blocks, objects

def build_prompt(ex_list, scenario_context):
    header = (
        "You are a planner. Return ONLY a valid JSON object with keys:\n"
        '- "complexity_level": integer\n'
        '- "target_action_sequence": array of strings\n\n'
        "Rules:\n"
        "- Actions MUST be in the exact parenthesized format shown in the examples.\n"
        "- Do not add explanations, markdown, or extra keys.\n"
        '- "complexity_level" MUST equal len(target_action_sequence).\n\n'
    )

    header += "\nIMPORTANT: Output MUST start with '{' and end with '}'.\n"

    shots = []
    for ex in ex_list:
        shots.append(
            "EXAMPLE SCENARIO:\n"
            f"{ex['scenario_context']}\n\n"
            "EXAMPLE OUTPUT:\n"
            + json.dumps(
                {
                    "complexity_level": ex["complexity_level"],
                    "target_action_sequence": ex["target_action_sequence"],
                },
                ensure_ascii=False
            )
            + "\n"
        )

    query = (
        "\nNOW SOLVE THIS SCENARIO:\n"
        f"{scenario_context}\n\n"
        "OUTPUT JSON ONLY:\n"
    )

    return header + "\n".join(shots) + query

def generate_json(model, tokenizer, prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    out = model.generate(
        **inputs,
        do_sample=False,
        temperature=0.0,
        max_new_tokens=256,
    )
    txt = tokenizer.decode(out[0], skip_special_tokens=True)

    start = txt.rfind("{")
    end = txt.rfind("}")
    parsed = json.loads(txt[start:end+1])

    seq = parsed["target_action_sequence"]
    parsed["complexity_level"] = int(len(seq))
    return parsed

def main_por_id(model, tokenizer, target_id):
    examples = load_json(EXAMPLES_PATH)
    tasks = load_json(TASK_PATH)

    blocks_ex, objects_ex = split_examples(examples)

    blocks_shots = blocks_ex[:3]
    objects_shots = objects_ex[:3]

    t = next(x for x in tasks if x["assembly_task_id"] == target_id)

    sc = t["scenario_context"]
    shots = blocks_shots if "set of blocks" in sc else objects_shots

    prompt = build_prompt(shots, sc)
    pred = generate_json(model, tokenizer, prompt)

    result = {
        "assembly_task_id": t["assembly_task_id"],
        **pred
    }

    print(json.dumps(result, indent=2, ensure_ascii=False))

def main(model, tokenizer):
    examples = load_json(EXAMPLES_PATH)
    tasks = load_json(TASK_PATH)

    blocks_ex, objects_ex = split_examples(examples)

    blocks_shots = blocks_ex[:3]
    objects_shots = objects_ex[:3]

    outputs = []

    for t in tasks:
        scenario_context = t["scenario_context"]

        if "set of blocks" in scenario_context:
            shots = blocks_shots
        else:
            shots = objects_shots

        prompt = build_prompt(shots, scenario_context)

        pred = generate_json(model, tokenizer, prompt)

        seq = pred["target_action_sequence"]
        result = {
            "assembly_task_id": t["assembly_task_id"],
            "complexity_level": len(seq),
            "target_action_sequence": seq
        }

        outputs.append(result)

    with open("submission.json", "w", encoding="utf-8") as f:
        json.dump(outputs, f, indent=2, ensure_ascii=False)

    print("‚úÖ submission.json generado correctamente")
```
